```python
# Import dependencies
import pandas as pd
from datetime import datetime

from birdnetlib import Recording
from birdnetlib.analyzer import Analyzer
from birdnetlib.analyzer_lite import LiteAnalyzer
from birdnetlib.batch import DirectoryAnalyzer
```

```python
# Load and initialize the BirdNET-Analyzer models.
analyzer = Analyzer()

recording = Recording(
    analyzer,
    "S4A1_20220310_040700.wav",
    lat=35.4244,
    lon=-120.7463,
    date=datetime(year=2022, month=5, day=10), # use date or week_48
    min_conf=0.25,
)
recording.analyze()
print(recording.detections)
```

```python
def on_analyze_complete(recording):
    print(recording.path)
    print(recording.detections)

directory = DirectoryAnalyzer(
    "C:\\Users\\Soldo\\Desktop\\BirdNet",
    patterns=["*.mp3", "*.wav"]
)
directory.on_analyze_complete = on_analyze_complete
directory.process()
```

```python
# initialize an empty dataframe to store detections
detections_df = pd.DataFrame(columns=["file", "common_name", "scientific_name", "start_time", "end_time", "confidence"])

# define a function to add detections to the dataframe
def add_detections_to_df(recording):
    global detections_df
    for detection in recording.detections:
        detection_dict = {
            "file": recording.path,
            "common_name": detection["common_name"],
            "scientific_name": detection["scientific_name"],
            "start_time": detection["start_time"],
            "end_time": detection["end_time"],
            "confidence": detection["confidence"],
        }
        detections_df = detections_df.append(detection_dict, ignore_index=True)

# configure directory analyzer
directory = DirectoryAnalyzer(
    "C:\\Users\\Soldo\\Desktop\\BirdNet",
    patterns=["*.mp3", "*.wav"],
)
directory.on_analyze_complete = on_analyze_complete

# process recordings
directory.on_analyze_complete = add_detections_to_df
directory.process()

# print the resulting dataframe
print(detections_df)
```
